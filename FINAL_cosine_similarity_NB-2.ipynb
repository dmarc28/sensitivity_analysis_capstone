```python
import re

def extract_numeric(df, col):
    df[col] = df[col].astype(str)  # Convert the column to string type
    pattern = r'\d+\.?\d*'  # Regular expression pattern to match numbers (including decimals)

    def extract_number(value):
        match = re.search(pattern, value)
        if match:
            return match.group()
        return None

    return df[col].apply(extract_number)
```
