# Implimentation of roBERTa pre-trained model

Let's import AutoTokenizer, AutoModelForSequenceClassification from transformers library of Hugging Face. AutoTokenizer automatically choses an appropriate tokenizer and AutoModelForSequenceClassification help loading pretrained models. Softmax from Scipy.special provide a probability distribution.

the implimentation of roberta model from Hugging Face can be found here: https://huggingface.co/cardiffnlp/twitter-roberta-base-sentiment

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

**Let's import the model: f"cardiffnlp/twitter-roberta-base
Instantiate the AutoTokenizer and the model**

```python
task='sentiment'
MODEL = f"cardiffnlp/twitter-roberta-base-{task}"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

## Now let's write a function that will return the dominant sentiment. 
1. tokenize the text and get a return as pytorch Tensors.
2. feed the encoded text in the model and save the output as output.
3. convert the tensor output from the model and save it as numpy array.
4. finally convert the score to get the score in range of 0 to 1.
score[0] = "negative"
score[1] = "neutral"
score[2] = "positive"

**For this project I want the dominant sentiment so I will write a if statement that returns only the either negative, neutral, positive.**

```python
def roberta_sentiment_test(text):
    try:
        encoded_text = tokenizer(text, return_tensors='pt')
        output = model(**encoded_text)
        scores = output[0][0].detach().numpy()
        scores = softmax(scores)

        if scores[0] > scores[1] and scores[0] > scores[2]:
            return ("negative")
        elif scores[2] > scores[1] and scores[2] > scores[0]:
            return ("positive")
        else:
            return("neutral")
    except: 
        ValueError
        return("nan")
        
```

# Let's write another function that will allow me to run the roberta_sentiment_test() function in chunks.

```python
roberta_sntvt_analyses_2 = {}
```

```python
def run_sensitivity(range1, range2):
    for i, row in df.iterrows():
        if i <= range2 and i >= range1:
            text = row['clean_text']
            set_id = row['set_id']
            roberta_sntvt_analyses_2[set_id] = roberta_sentiment_test(text)
```

The following function generate a csv file from the dataframe and generates download link.
This code is from another Kaggle notebook "https://www.kaggle.com/code/rtatman/download-a-csv-file-from-a-kernel"

```python
# import the modules we'll need
from IPython.display import HTML
import pandas as pd
import numpy as np
import base64

# function that takes in a dataframe and creates a text link to  
# download it (will only work for files < 2MB or so)
def create_download_link(dd_f, title = "Download CSV file", filename = "twtr_b1.csv"):  
    csv = dd_f.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

# create a random sample dataframe
download_df = pd.DataFrame(np.random.randn(50, 4), columns=list('ABCD'))

```
