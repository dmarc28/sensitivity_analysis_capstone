```python
import re

def clean_text(text):
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Remove Twitter handles
    text = re.sub(r'@\w+', '', text)
    
    # Remove hashtags
    text = re.sub(r'#\w+', '', text)
    
    # Remove special characters and symbols
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove non-ASCII characters
    text = text.encode('ascii', 'ignore').decode('utf-8')
    
    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
df["clean_text"] = df["text"].apply(clean_text)
```

```python
import nltk
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
custom_stop_words = ['via', "doesn't", "thi", "that", "whi", "dont", "govern", "offici", "charg", "investig", "say", "get", "new"]
```

```python

def remove_stop_words(text):

    # Tokenize the text into individual words
    words = nltk.word_tokenize(text)

    # Remove stop words from the list of words
    filtered_words = [word for word in words if word.lower() not in stop_words and word.lower() not in custom_stop_words]

    # Reconstruct the filtered text
    filtered_text = ' '.join(filtered_words)

    # Print the filtered text
    return(filtered_text)
```

```python
df["more_clean_text"] = df["clean_text"].apply(remove_stop_words)
```

```python
def classify_text(document):
    blob = TextBlob(document)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score < - 0.2:
        return "negative"
    elif sentiment_score > 0.5:
        return "positive"
    else:
        return "neutral"
```

```python
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize
```

```python

def lemmatize_and_stem(text):
    try:
        # Lemmatize the text
        lemmatizer = WordNetLemmatizer()
        lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokenize(text)]
        lemmatized_text = ' '.join(lemmatized_words)

        # Apply Stemming to the lemmatized text
        stemmer = PorterStemmer()
        stemmed_words = [stemmer.stem(word) for word in word_tokenize(lemmatized_text)]
        stemmed_text = ' '.join(stemmed_words)

        return stemmed_text
    except:
        TypeError
        return "nan"

```
